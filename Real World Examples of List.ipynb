Real World Examples using Lists in Python

Lists are one of the most commonly used data structures in Python, thanks to their versatility and ease of use. Here are several practical examples that illustrate their use in real-world scenarios.

Example 1. Manage a To Do List

```python
to_do_list = ["Buy Groceries", "Clean the House", "Pay Bills"]

##Adding to task
to_do_list.append("Walk the Dog")
to_do_list.append("Read a Book")

##Removing a task
to_do_list.remove("Pay Bills")

#Checking if a task is in the list
if "Clean the House" in to_do_list:
    print("Task is in the list.")

print("To Do List remaining")
for task in to_do_list:
    print(f"-{task}")
```

Example 2: Organizing Student Grades

```python
grades = [85, 92, 78, 90, 88]

#Added a new grade
grades.append(95)

#Calculate average grade
average_grade = sum(grades) / len(grades)
print(f"Average Grade: {average_grade: .2f}")

#Find highest and lowest grades
highest_grade = max(grades)
lowest_grade = min(grades)
print(f"Highest Grade: {highest_grade}")
print(f"Lowest Grade: {lowest_grade}")
```

Example 3: Managing an Inventory

```python
inventory = ["Apples", "Bananas", "Oranges", "Grapes"]

#Adding new items to inventory
inventory.append("Mangoes")

#Removing an item from inventory
inventory.remove("Bananas")

#Check if an item is in stock
item = "Grapes"
if item in inventory:
    print(f"{item} is in stock.")
else:
    print(f"{item} is out of stock.")


print("Current Inventory:")
for fruit in inventory:
    print(f"- {fruit}")
```

Example 4 : Collecting User Feedback

```python
feedback = ["Great service!", "Could improve delivery time.", "Excellent product quality.", "Very Satisfied!"]

#Adding new feedback

feedback.append("Will shop again.")

#Counting specific feedback

positive_feedback_count = sum(1 for comment in feedback if "Great" in comment or "Excellent" in comment or "Satisfied" in comment)
print(f"Positive Feedback Count: {positive_feedback_count}")

print("All User Feedback:")
for comment in feedback:
    print(f"- {comment}")
```

Example 5 : Check if a String is Palindrome

```python
def is_palindrome(s):
    s = s.lower().replace(" ", "")
    return s==s[::-1]

print(is_palindrome("A man a plan a canal Panama"))
print(is_palindrome("hello"))
```
