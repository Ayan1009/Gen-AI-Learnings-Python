```python
## Creating Dictionaries
empty_dict = {}
print(empty_dict)
```

```python
empty_dict = dict()
print(empty_dict)
```

```python
student={"name":"Ayan", "age":26, "grade":"A"}  ##Key Value pairs
print(student)
print(type(student))
```

```python
# Single key cannot have multiple values, the last value will be assigned to the key
student={"name":"Ayan", "age":26, "name":"A"}
print(student)
```

```python
#Accessing Dictionary Elements
student={"name":"Ayan", "age":26, "grade":"A"}
print(student["name"])
print(student["age"])

#Accessing using get() method
print(student.get("grade"))
print(student.get("address"))  # Returns None if key not found
print(student.get("address", "Not Found"))  # Returns default value if key not found
```

```python
#Modifying Dictionary Elements
# Dictionaries are mutable, so we can change their content that is add, update or delete key-value pairs.

student["age"] = 27  # Update existing key with a new value
student["address"] = "123 Main St"  # Add new key-value pair

print(student)

del student["grade"]  # Remove key-value pair by key
print(student)
```

```python
# Dictionary Methods

keys = student.keys()  # Get all keys
print(keys)
values = student.values()  # Get all values
print(values)

items = student.items()  # Get all key-value pairs
print(items)

#Key value pairs are always in the form of tuples
```

```python
#Shallow Copy

student_copy = student  # Create a shallow copy of the dictionary
print(student)
print(student_copy)
```

```python
student["name"] = "Ayan2"  # Modify original dictionary
print(student)
print(student_copy)  # Copy remains unchanged
```

```python
student_copy1 = student.copy()  # Create a shallow copy of the dictionary
#Shallow copy basically means that the reference is copied not the actual object.
#It will take a different memory location and then the reference of this variable will be copied to the new variable.


print(student)
print(student_copy1)
```

```python
#Example of Shallow Copy

student["name"] = "Ayan3"
print(student_copy1)
print(student)
```

```python
# Iterating over Dictionaries
## You can use loops to iterate over the keys, values, or key-value pairs or items in a dictionary.

for key in student.keys():
    print(key)

for value in student.values():
    print(value)

for key, value in student.items():
    print(f"{key}: {value}")
```

```python
### Nested Dictionaries

students = {
    "student1": {"name": "Ayan", "age": 26},
    "student2": {"name": "Wonder Woman", "age": 19}
}
print(students)
```

```python
## Access nested dictionary elements

print(students["student1"]["name"])  
print(students["student2"]["age"])   
```

```python
## Iterating over Nested Dictionaries

for student_id, info in students.items():
    print(f"{student_id}:{info}")
    for key, value in info.items():
        print(f"{key}: {value}")
```

```python
## Dictionary Comprehensions

squares = {x: x**2 for x in range(6)} ## Over here x is the key and x**2 is the value
print(squares)
```

```python
## Conditional Dictionary Comprehensions

even_squares = {x: x**2 for x in range(10) if x % 2 == 0}
print(even_squares)
```

```python
## Practical Example: Counting Frequency of Elements in a List

numbers = [1, 2, 2, 3, 4, 4, 4, 5]
freq = {}

for num in numbers:
    if num in freq:
        freq[num] = freq[num] + 1
    else:
        freq[num] = 1
print(freq)
```

```python
## Merge 2 Dictionaries into one

dict1 = {"a":1, "b":2}
dict2 = {"c":3, "d":4}
merged_dict = {**dict1, **dict2}  # dict2 values will overwrite dict1 values for duplicate keys
print(merged_dict)
```
