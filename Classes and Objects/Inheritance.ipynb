Inheritance in Python

Inheritance is a fundamental concept in Object-Oriented Programming (OOP) that allows a class to inherit
attributes and methods from another class. This lesson covers single inheritance and multiple inheritance,
onstrating how to create and use them in Python.

```python
### Inheritance
## Parent Class

class Car:
    def __init__(self, windows, doors, enginetype):
        self.windows = windows
        self.doors = doors
        self.enginetype = enginetype

    def drive(self):
        print(f"The person will drive the {self.enginetype} car")
```

```python
car1 = Car(4, 4, "petrol")
car1.drive()
```

```python
class Tesla(Car):
    def __init__(self, windows, doors, enginetype, autopilot):
        super().__init__(windows, doors, enginetype)   ### Super() is used to call the parent class constructor
        self.autopilot = autopilot

    def self_drive(self):
        print(f"The person will drive the {self.enginetype} car with autopilot {self.autopilot}")
```

```python
car2=Tesla(4,4,"electric","full self-driving")
car2.drive()
```

```python
### Multiple Inheritance
## When a class inherits from more than one parent class, it is called multiple inheritance.

## Base class 1
class Animal:
    def __init__(self, name):
        self.name=name

    def speak(self):
        print("Subclass must implement this method")

## Base class 2

class Pet:
    def __init__(self, owner):
        self.owner=owner

## Derived class
class Dog(Animal, Pet):
    def __init__(self, name, owner):
        Animal.__init__(self,name)
        Pet.__init__(self,owner)

    def speak(self):
        print(f"{self.name} says Woof! and is owned by {self.owner}")


dog = Dog("Buddy", "Alice")
dog.speak()
```

```python
### Modeling a real-world scenario using multiple inheritance

## Base class 1
class Employee:
    def __init__(self, name, id):
        self.name = name
        self.id = id

    def work(self):
        print(f"{self.name} is working with employee ID {self.id}.")

## Base class 2
class Manager:
    def __init__(self, department):
        self.department = department

    def manage(self):
        print(f"Managing the {self.department} department.")

## Derived class
class TeamLead(Employee, Manager):
    def __init__(self, name, id, department):
        Employee.__init__(self, name, id)
        Manager.__init__(self, department)

    def lead(self):
        print(f"{self.name} is leading the team in the {self.department} department.")
```

```python
employee = TeamLead("John Doe", 12345, "Software Development")
employee.work()
employee.manage()   
employee.lead()
```
