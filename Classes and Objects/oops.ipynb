Classes and Objects

Object-Oriented Programming (OOP) is a programming paradigm that uses "objects" to design applications and computer programs. OOP allows for modeling real-world scenarios using classes and objects. This lesson covers the basics of creating classes and objects, including instance variables and methods.

```python
### A class is a blueprint for creating objects. It defines a set of attributes and methods that the objects created from the class will have. In Python, you can define a class using the `class` keyword followed by the name of the class and a colon

class Car:
    pass

audi = Car()
bmw = Car()

audi.windows = 4

print(type(audi))
print(audi.windows)
```

```python
print(audi)
print(bmw)
```

```python
tata = Car()

tata.doors = 4
print(tata.windows) ## This will raise an AttributeError because the 'windows' attribute is not defined for the 'tata' object.
```

```python
dir(tata) ## This will show the attributes and methods of the 'tata' object, which includes 'doors' but not 'windows'.
```

```python
### Instance Variable and Methods

class Dog:
    ## constructor
    def __init__(self, name, age):  ### name and age are parameters of the constructor or attributes of the class
        self.name = name   ### self.name is an instance variable that is assigned the value of the name parameter passed to the constructor when creating an object of the Dog class.
        self.age = age

## Creating an object of the Dog class

dog1 = Dog()
```

```python
dog1 = Dog("Buddy", 3)
print(dog1)
print(dog1.name)
print(dog1.age)

dog2 = Dog("Max", 5)
print(dog2.name)
print(dog2.age)
```

```python
### Define a class with Instance method

class Dog:
    ## constructor
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def bark(self):  ## instance method and it takes self as a parameter which refers to the instance of the class that is calling the method.
        print(f"{self.name} says Woof!")
    
    def silent_bark(self):
        print(f"{self.name} is silent.")

dog1 = Dog("Buddy", 3)
dog1.silent_bark() ## This will call the silent_bark method of the dog1 object, which will print "Buddy is silent." to the console.

dog2 = Dog("Max", 5)
dog2.bark() ## This will call the bark method of the dog2 object, which will print "Max says Woof!" to the console.
    
```

```python
## Modeling a bank account using a class

### Define a class for a bank account

class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print(f"{amount} deposited. New balance: {self.balance}")

    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient funds.")
        else:
            self.balance -= amount
            print(f"{amount} withdrawn. New balance: {self.balance}")

### Create an instance of the BankAccount class

account1 = BankAccount("Alice", 1000)
print(account1.owner)  ## This will print "Alice" to the console, which is the owner of the account.
print(account1.balance) ## This will print "1000" to the console, which is the initial balance of the account.
account1.deposit(500) ## This will call the deposit method of the account1 object, which will add 500 to the balance and print "500 deposited. New balance: 1500" to the console.
account1.withdraw(200) ## This will call the withdraw method of the account1 object, which will subtract 200 from the balance and print "200 withdrawn. New balance: 1300" to the console.
account1.withdraw(1500) ## This will call the withdraw method of the account1 object, which will check if 1500 is greater than the current balance (1300) and print "Insufficient funds." to the console since there are not enough funds in the account to complete the withdrawal.
```

```python
account2 = BankAccount("Bob", 10000)
print(account2.owner, account2.balance)  ## This will print "Bob" and "10000" to the console, which is the owner and initial balance of the account.
account2.deposit(2000) ## This will call the deposit method of the account2 object, which will add 2000 to the balance and print "2000 deposited. New balance: 12000" to the console.
account2.withdraw(9700)
account2.withdraw(3000)
```
