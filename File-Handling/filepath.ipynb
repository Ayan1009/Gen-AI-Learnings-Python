```python
### Using the os module

import os
cwd = os.getcwd()
print(f"Current working directory: {cwd}")
```

```python
## create a new directory

os.mkdir("new_directory")
print(f"Directory 'new_directory' created.")
```

```python
## Remove the created directory

os.rmdir("new_directory")
print(f"Directory 'new_directory' removed.")
```

```python
## List all the files and directories in the current directory

entries = os.listdir('.')
print("Entries in the current directory:")
for entry in entries:
    print(entry)
```

```python
## Joining paths

file_name = "example1.txt"
dir_name = "folder"
file_path = os.path.join(os.getcwd(), dir_name, file_name)

print(f"Full file path: {file_path}")
```

```python
path="example1.txt"
if os.path.exists(path):
    print(f"The file '{path}' exists.")
else:
    print(f"The file '{path}' does not exist.")
```

```python
with open("example1.txt", "w") as f:
    f.write("This is a new file created for testing.")

path = "example1.txt"
if os.path.exists(path):
    print(f"The file '{path}' exists.")
else:
    print(f"The file '{path}' does not exist.")
```

```python
import os
path = "new_directory"

if os.path.isfile(path):
    print(f"'{path}' is a file.")
elif os.path.isdir(path):
    print(f"'{path}' is a directory.")
else:
    print(f"'{path}' does not exist.")
```

```python
### Getting the absolute path of a file

relative_path = "example1.txt"
absolute_path = os.path.abspath(relative_path)
print(f"Absolute path of '{relative_path}': {absolute_path}")
```
