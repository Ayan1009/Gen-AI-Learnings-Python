File Operation - Read and Write Files


```python
### Read a whole file

with open('example.txt', 'r') as file:
    content=file.read()
    print(content)
```

```python
### Read a file line by line

with open('example.txt', 'r') as file:
    for line in file:
        print(line.strip())   ## strip() removes the newline character
        
```

```python
### Write a file(overwrite)

with open('example.txt', 'w') as file:
    file.write("This is a new line.\n")
    file.write("This line will overwrite the existing content.\n") ## Overwrites the file entirely
```

```python
### Write a file (without overwriting)

with open('example.txt', 'a') as file:
    file.write("This line will be appended to the file.\n") ## Appends to the file
```

```python
### Writing a list of lines to a file

lines=["First line\n", "Second line\n", "Third line\n"]
with open('example.txt', 'a') as file:
    file.writelines(lines)
```

```python
### Binary Files

## Writing binary data to a file

data = b'\x00\x01\x02\x03\x04\x05'
with open('example.bin', 'wb') as file:   ## wb means write binary
    file.write(data)
```

```python
### Binary Files

## Reading binary data from a file

with open('example.bin', 'rb') as file:   ## rb means read binary
    data = file.read()
    print(data)
```

```python
### Read the content from a source text file and write it to a new text file

with open("example.txt", "r") as file:
    content = file.read()

with open("Destination_file.txt", "w") as file:
    file.write(content)
```

```python
### Read a text file and count the number of lines, words, and characters

with open("example.txt", "r") as file:
    lines = file.readlines()
    line_count = len(lines)
    words = sum(len(line.split()) for line in lines)
    characters = sum(len(line) for line in lines)

    print(f"Lines count: {line_count}, Words count: {words}, Characters count: {characters}")
```

```python
### Writing and then reading a file

with open("example.txt", "w+") as file:
    file.write("Hello World!\n")
    file.write("Welcome to file handling in Python.\n")

    file.seek(0)  ## Move the file pointer to the beginning of the file
    content = file.read()
    print(content)
```
